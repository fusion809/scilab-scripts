```octave
N=1000;
zero=-13.69;
% Use Newton's method to find the 11th zero of the Airy Ai(x) function
while airy(0,zero)>1e-12
    zero=zero-airy(0,zero)/airy(1,zero);
end
% Start of the plotting domain
a=zero;
b=5;
% x
x=linspace(a,b,N+1)';
% Ai(x)
y=airy(0,x);
% Ai'(x)
yy=airy(1,x);
plot(x,y,'r',x,yy,'g')
```
